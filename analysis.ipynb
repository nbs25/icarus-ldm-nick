```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('./el_newdet.dat') as electron_file:
    electron_events = electron_file.readlines()
with open('./DM30MeV_newdet.dat') as ldm_file:
    ldm_events = ldm_file.readlines()
    
assert len(ldm_events) == len(electron_events)
```

```python
electron_px = []
electron_py = []
electron_pz = []
electron_E  = []
electron_x  = []
electron_y  = []
electron_z  = []

ldm_px = []
ldm_py = []
ldm_pz = []
ldm_E  = []
ldm_x  = []
ldm_y  = []
ldm_z  = []

electron_p = []

ldm_p = []

for i in range(min(len(electron_events),len(ldm_events))):
    
    ldm_event = [float(item) for item in ldm_events[i].strip().split('\t')]
    electron_event = [float(item) for item in electron_events[i].strip().split('\t')]
        
    assert len(ldm_event) == len(electron_event)
    
    electron_px.append(electron_event[0])
    electron_py.append(electron_event[1])
    electron_pz.append(electron_event[2])
    electron_E.append(electron_event[3])
    electron_x.append(electron_event[4])
    electron_y.append(electron_event[5])
    electron_z.append(electron_event[6])
    
    ldm_px.append(ldm_event[0])
    ldm_py.append(ldm_event[1])
    ldm_pz.append(ldm_event[2])
    ldm_E.append(ldm_event[3])
    ldm_x.append(ldm_event[4])
    ldm_y.append(ldm_event[5])
    ldm_z.append(ldm_event[6])
    
    electron_p.append(np.sqrt(electron_event[0]**2 + electron_event[1]**2 + electron_event[2]**2 ))
    ldm_p.append(np.sqrt(ldm_event[0]**2 + ldm_event[1]**2 + ldm_event[2]**2 ))
```

```python
bins=np.histogram(np.hstack((electron_px,ldm_px)), bins=40)[1]
n,bins,patches = plt.hist(electron_px, bins, facecolor='blue',alpha=0.5,label='$e^{-}$')
n,bins,patches = plt.hist(ldm_px, bins, facecolor='red',alpha=0.5,label='$ldm_{30}$')
plt.xlabel("$p_{x}$ [GeV]")
plt.ylabel("count [unitless]")
plt.legend()
plt.show()
```

```python
bins=np.histogram(np.hstack((electron_py,ldm_py)), bins=40)[1]
n,bins,patches = plt.hist(electron_py, bins, facecolor='blue',alpha=0.5,label='$e^{-}$')
n,bins,patches = plt.hist(ldm_py, bins, facecolor='red',alpha=0.5,label='$ldm_{30}$')
plt.xlabel("$p_{y}$ [GeV]")
plt.ylabel("count [unitless]")
plt.legend()
plt.show()
```
